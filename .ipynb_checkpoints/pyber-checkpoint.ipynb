```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# loading data files
city_data_to_load = "./data/city_data.csv"
ride_data_to_load = "./data/ride_data.csv"

# Read the City and Ride Data
city=pd.read_csv(city_data_to_load)
ride=pd.read_csv(ride_data_to_load)
```

```python
# Combine the data into a single dataset
city_ride=pd.merge(ride,city,on="city",how="outer")

# Display the data table for preview
city_ride.head()
```

```python
# Creating series for ride counts, average fare, and driver counts by city/type
city_stats=city_ride.groupby(["city","type"])
ride_count=city_stats["ride_id"].count()
mean_fare=city_stats["fare"].mean()
driver_count=city_stats["driver_count"].last()
```

## Bubble Plot of Ride Sharing Data

```python
# Creating ride count series for x-axis by city type
x_urban=ride_count[ride_count.index.get_level_values("type")=="Urban"]
x_suburban=ride_count[ride_count.index.get_level_values("type")=="Suburban"]
x_rural=ride_count[ride_count.index.get_level_values("type")=="Rural"]

# Creating mean fare series for y-axis by city type
y_urban=mean_fare[mean_fare.index.get_level_values("type")=="Urban"]
y_suburban=mean_fare[mean_fare.index.get_level_values("type")=="Suburban"]
y_rural=mean_fare[mean_fare.index.get_level_values("type")=="Rural"]

# Build the scatter plots for each city types
plt.scatter(x_urban, y_urban, marker="o", facecolors="gold", edgecolors="black",
            s=driver_count[driver_count.index.get_level_values("type")=="Urban"], 
            alpha=.5, linewidths=1)
plt.scatter(x_suburban, y_suburban, marker="^", facecolors="lightskyblue",edgecolors="black",
            s=driver_count[driver_count.index.get_level_values("type")=="Suburban"], 
            alpha=.5, linewidths=1)
plt.scatter(x_rural, y_rural, marker="D", facecolors="coral", edgecolors="black",
            s=driver_count[driver_count.index.get_level_values("type")=="Rural"], 
            alpha=.5, linewidths=1)

#plot title, axes labels
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

# Creates a legend
plt.legend(['Urban','Suburban','Rural'], loc='best')
# Incorporates a text label regarding circle size
plt.text(15,42,'Note:\nThe circle size correlates with\ndriver count per city',size=7)


# Save Figure
plt.savefig("./Images/bubble_plot.png",dpi=256,figsize=(18,11))

# show plot
plt.show()
```

## Total Fares by City Type

```python
# order of city types
city_types=["Urban","Suburban","Rural"]
# fares by city type ordered as in 'city_types' list
city_type_fares=city_ride.groupby(["type"])["fare"].sum().reindex(city_types)
# exploding the biggest city type in terms of total fare, "Urban"
explode=[.1,0,0]
```

```python
# generating pie chart
# percentages will be automatically populated
plt.pie(city_type_fares, explode=explode, labels=city_types, colors= ["gold","lightskyblue","coral"],
        autopct="%1.1f%%", shadow=True, startangle=140)

# title
plt.title("% of Total Fares by City Type")

# saving figure; dpi for resolution, figsize for printing consideration
plt.savefig("./Images/fares_pie.png",dpi=256,figsize=(18,11))

# show plot
plt.show()
```

## Total Rides by City Type

```python
# counting rides by city type and ordering index as in 'city_types'
city_type_rides=city_ride.groupby(["type"])["ride_id"].count().reindex(city_types)
```

```python
# generating pie chart
plt.pie(city_type_rides, explode=explode, labels=city_types, colors= ["gold","lightskyblue","coral"],
        autopct="%1.1f%%", shadow=True, startangle=140)
# setting title
plt.title("% of Total Rides by City Type")

# saving figure
plt.savefig("./Images/rides_pie.png",dpi=256,figsize=(18,11))

# show plot
plt.show()
```

## Total Drivers by City Type

```python
# driver_counts were duplicated during initial merge between "city" and "ride"
# in order to capture unique driver_count by city we use groupby twice
city_type_drivers=city_ride.groupby(["city","type"],as_index=False)["driver_count"].last().groupby(["type"])["driver_count"].sum().reindex(city_types)
```

```python

# Build Pie Charts
plt.pie(city_type_drivers, explode=explode, labels=city_types, colors= ["gold","lightskyblue","coral"],
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Drivers by City Type")
#plt.show()
# Save Figure
plt.savefig("./Images/drivers_pie.png",dpi=256,figsize=(18,11))

# show plot
plt.show()
```

# Trends:
1. Urban sector dominates other sectors in terms of total fares, number of rides, and number of drivers
2. There is a negative correlation between the number of rides and average fare in a city, indicating possibly thinner profit margins in busier markets. Expanding business in rural, and suburban markets should be considered
3. Although urban sector produces the majority of fare revenue, it is not leading in terms of fare generated per driver, indicating the possibility of an abundance of drivers in the urban market not being met by relatively sufficient riding opportunities, or more fierce competition in the urban market
